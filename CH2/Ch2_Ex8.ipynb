```python
# 2 Rename the Unnamed: 0 column to College and set the index as the latter

df_college3 = df_college.rename({'Unnamed: 0' : 'College'}, axis = 1) 

df_college3 = df_college3.set_index('College')

df_college3.head()
```

### e)

```python
fig, ax = plt.subplots(figsize=(10, 8))

df_college.boxplot('Outstate', by = 'Private', ax = ax)

ax.set_xlabel('Private')
ax.set_ylabel('Outstate Tuition numbers')
ax.set_title('Outstate Tuition numbers grouped by Private')
```

- Private institutions provide more out of state tuitions 

### f) Create a new column categorizing if an Institution is elite or not, based on the criterion that if the number of top 10 percent high school graduates in that institution is more than 50

```python
df_college['Elite'] = df_college['Top10perc'].apply(lambda x: 'Yes' if x > 50 else 'No')

df_college['Elite'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

df_college.boxplot('Outstate', by = 'Elite', ax = ax)

ax.set_xlabel('Institution Category')
ax.set_ylabel('Out-of-state Tuition rate')
ax.set_title('Out-of-state Tuition rate grouped by Eliteness of the Institution')
```
