a)

```python
flex = np.arange(0., 10.0, 0.01)

def sq_bias(x):
    return 2.4 * (10**(-3)) * (-x+10)**(3)
def var(x):
    return 0.1 + .002*x**3 
def train_error(x): # A function that decreases monotonically as its argument increase
    return 2.5 - 0.6*x + 0.13*x**2 - 0.0123*x**3 + 0.0003*x**4
def test_error(x):
    return sq_bias(x) + var(x) + 1. # Bayes error is set to 1

plt.figure(figsize=(8, 6))

plt.ylim(0,3.5)

plt.plot(flex,sq_bias(flex), label='squared bias')
plt.plot(flex, var(flex), label='variance')
plt.plot(flex, train_error(flex), label='training error')
plt.plot(flex, test_error(flex), label='test error')

plt.axhline(1, color = 'grey', linestyle = 'dashed')


plt.legend(loc='best')
plt.xlabel(r'$\textrm{model flexibility}$')

plt.show()
```

2.3) 

b)

Squared bias: This is the error introduced by approximating a real life phenomenon by an approximate function hat{f}. Increasing the flexibility of the model generically leads to a decrease in squared bias and in theory we
can reduce it to zero. 

Variance: Less flexible models do not vary much when they see new data, e.g a linear model will shift less in slope. More flexible model will tend to have more variance when they encounter new data. 

Training error: Training error will reduce always, when we increase the degrees of freedom in a model as the algorithm picks parameters/weights that reduce the mean squared error between the actual label and the label 
predicted by the model. 

Test error: Variance + Bias + Bayes error (or var(noise)) and hence obtain a convex shape. Since all the error is positive it is always positive. 

Bayes error. This term is the irreducible error/noise and assumed to be constant on average because it does not 
depend on the features X and hence do not depend on the model.

5) 
Advantages of 
Flexible models with respect to less flexible ones: Low bias, given enough sample size better results in general. 
Disadvantages: Less interpretibility, prone to over-fitting problems, computationally expensive (longer training time etc.)

Flexible models can be preferred when i) Sample size is large ii) when non-linear complex relationships are expected between predictors and the target

On the other hand, less flexible models can be preffered i) When the sample size is small with lots of predictors
ii) when the data has large variance, i.e with large error terms. 

6) 

The advantages of a parametric methods are: requires less observations, are faster and computational less expensive and interpretible in the sense that one can find rough relationships between predictors and target. On the other hand, these relationships are often approximate due to the constraints on the parametric approaches which might cause them to miss more intricate relationships in the data.
