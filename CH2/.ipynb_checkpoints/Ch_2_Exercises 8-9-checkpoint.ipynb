### 8. We begin by analysing College.csv data set, which contains a number of predictors for 777 different universities and colleges in the USA. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib as mpl

# For inline plotting 

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

# Plot parameters defined globally
mpl.rcParams['text.usetex'] = True

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.size'] = 2
plt.rcParams['xtick.major.width'] = 1
plt.rcParams['ytick.major.size'] = 2
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.right'] = True
```

```python
college = pd.read_csv('College.csv') # Load the data csv file and call it 'college'
```

```python
college.head() #First five rows of the dataframe
```

Notice that there is a column labeling names of the colleges. We do not want to treat this column as a data and we can do so in various ways 

```python
# 1, Load the data frame by dropping the index column, which is Unnamed:0
college2 = pd.read_csv('College.csv', index_col=0)
```

```python
college2.head() # The resulting dataframe has the first column 'Private'
```

```python
# 2 Rename the Unnamed: 0 column to College and set the index as the latter

college3 = college.rename({'Unnamed: 0' : 'College'}, axis = 1) 

college3 = college3.set_index('College')

college3.head()
```

```python
college = college3
```

```python
college.describe()
```

```python
college[['Top10perc','Apps','Enroll']]
```

```python
pd.plotting.scatter_matrix(college[['Top10perc','Apps','Enroll']], figsize = (10,10))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

college.boxplot('Outstate', by = 'Private', ax = ax)

ax.set_xlabel('Private')
ax.set_ylabel('Outstate Tuition numbers')
ax.set_title('Outstate Tuition numbers grouped by Private')
```

```python
# Private institutions provide more out of state tuitions 
```

```python
college['Top10perc']
```

```python
# Create a new column categorizing if an Institution is elite or not, based on the criterion that if the number
# of top 10 percent high school graduates in that institution is more than 50
```

```python
college['Elite'] = college['Top10perc'].apply(lambda x: 'Yes' if x > 50 else 'No')
```

```python
college['Elite']
```

```python
college['Elite'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

college.boxplot('Outstate', by = 'Elite', ax = ax)

ax.set_xlabel('Institution Category')
ax.set_ylabel('Out-of-state Tuition rate')
ax.set_title('Out-of-state Tuition rate grouped by Eliteness of the Institution')
```

```python
college.head()
```

```python
# make some histogram plots of quantitative variables 

fig, axes = plt.subplots(2,2,figsize = (10,10))

college.hist(column = 'PhD', grid = False, bins = 20, ax = axes[0,0], color = 'green')
college.hist(column = 'S.F.Ratio', grid = False, bins = 20, ax = axes[0,1], color = 'red')
college.hist(column = 'perc.alumni', grid = False, bins = 30, ax = axes[1,0], color = 'blue')
college.hist(column = 'Expend', grid = False, bins = 40, ax = axes[1,1], color = 'black')
```

```python
college.hist(column = 'S.F.Ratio', by = 'Private', grid = False, bins = 20, color = ['red'])
```

- Student to faculty ratio is tend to be larger amoung public institutions 

```python
college.hist(column = 'S.F.Ratio', by = 'Elite', grid = False, bins = 15, color = ['red'])
```

- For Elite institutions, student to Faculty ratio is always less than 20 percent, this distribution is not Gaussian.

```python
college.hist(column = 'Grad.Rate', by = 'Private', grid = False, bins = 15, color = ['orange'])
```

- Among the Private inst., mean graduation rate is around 75 percent higher than public ones

```python
plt.scatter(college['Outstate'],college['Grad.Rate'], marker = '+', color = 'darkblue')
```

```python
# Seems like a corrupt data as there is more than 100 percent Grad rate for one institution, the reason for this
# could be that they have taken account students with double affiliation for example
```

- The plot above also clarifies one trend high tuition fees correlates with high graduation rate!!! 

```python
plt.scatter(college['PhD'],college['Outstate'], marker = 'x')
```

-  Institutions with high percentage of the faculty with Ph.D. degree tends to correlate with high tuition fees! 

```python
plt.scatter(college['Accept']/college['Apps'],college['S.F.Ratio'], marker = 'o', color = 'red')
```

-  Institutes with high acceptence rate has higher Student to Faculty ratio! 

```python
plt.scatter(college['Outstate'],college['Expend'], marker = 'o', color = 'red')
```

-  Institutes with high tuition fees can spend more on the instructional expenditure per student

```python
plt.scatter(college['Top10perc'],college['Grad.Rate'], marker = 'o', color = 'blue')
```

- Institutions with a low number of students from the top 10 percent of their high school can also have a high graduation rate. This is clear from the fact that most of the institutions that have low number of students with that attribute still have relatively large graduation rate. Therefore, institutions with high number of students in the first 10 percentile of their high schools does not indicate high graduation rate!
