### d)

```python
def Power3(x,a):

    return x**a
```

### e) - f)

```python
def plotpowers(xin, xend, a, log_scale = None):
    
    fig, axes = plt.subplots(figsize = (10,6))
    
    x = np.arange(xin,xend)
    
    y = Power3(x,a)
    
    axes.set_title('Powers of x')

    axes.set_xlabel('x')
    axes.set_ylabel(f'$x^{a}$')    
    
    if log_scale == 'x':

        axes.set_xscale('log')
        axes.set_xlabel('log(x)')

    if log_scale == 'y':
    
        axes.set_yscale('log')
        axes.set_ylabel(f'$log(x^{a})$')
    
    if log_scale == 'xy':
        
        axes.set_xscale('log')
        axes.set_yscale('log')
        
        axes.set_xlabel('log(x)')
        axes.set_ylabel(f'$log(x^{a})$')
        
    axes.plot(x,y)
```

```python
plotpowers(1,11,5)
```

```python
plotpowers(1,11,5, log_scale='xy')
```
