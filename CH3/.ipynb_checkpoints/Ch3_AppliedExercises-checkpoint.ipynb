### Exercise 3.8

#### 8.a - Perform a linear regression using horsepower as a predictor and mpg as the response (targret), summarize the results and comment.

```python
X_train = sm.add_constant(X_train) # to work with sm.OLS we apparently need to add an intercept via add_constant
```

```python
X_train.head()
```

```python
model = sm.OLS(y_train, X_train)
```

```python
fit_results = model.fit()
```

```python
print(fit_results.summary())
```

### Comments: 

- i. There is a clear relationship between the horsepower and the mpg, p value of the predictor is very low, F-statistic is much larger than 1, its std is small and the confidence interval of horsepower does not contain 0: [-0.171, -0.145]. We can confidently reject the null hypothesis that the regression coefficient $\beta_j$ of horsepower is zero.

- ii. As $R^2 = 0.606$, we can state that about 60 percent of the variance in the response (mpg) can be explained by the horsepower. 

- iii. The relationship between horsepower vs mpg is negative as the regression coefficient $\beta_j = - 0.1578$

- iv. According to the analysis below, the prediction for 98 horsepower is mpg = 24.467 with 95 percent confidence interval = [23.97, 24.96] and prediction interval = [14.81,34.12]

```python
fit_results.params[1]
```

```python
predict_df = pd.DataFrame({'constant': [1.0] ,'horsepower': [98.0]})
```

```python
predict_df.head()
```

```python
predictions = fit_results.get_prediction(predict_df)

print(f' Predicted value of Miles per Gallon for 98 horsepower is: {predictions.predicted_mean[0]}')
```

```python
print(f' 95 % Confidence interval is : {predictions.conf_int(alpha=0.05)[0]}')
```

```python
print(f' 95 % Prediction interval is : {predictions.conf_int(obs = True, alpha=0.05)[0]}')
```

Recall that assuming the true population is captured by a linear model: 

$$ f(X_1) = \beta_0 + \beta_1 X_1 $$

we have estimated the coefficients $\hat{\beta}_0$ and $\hat{\beta}_1$ to get a prediction $\hat{Y} = \hat{\beta}_0 + \hat{\beta}_1 X_1$. In this way, we average over an ensemble of car data assuming the random irredicuble error $\epsilon_i$ is canceled through the averagin process. Confidence interval is a measure of how accurate our prediction $\hat{Y}$ is close to actual ensemble average response $f(X_1)$. Prediction interval on the other hand, quantifies the error we make for an individual observation including the irreducible error $\epsilon$ in the model 
$ f(X_1) = \beta_0 + \beta_1 X_1 + \epsilon$. Therefore the $95 \%$ interval for the latter is much wider although they are both centered around the mean estimate. 

#### 8.b - Plot the response and the predictor together with the least square regression line

```python
# First we define a function that adds a least square line to an axis defined 
```

```python
def lsline(axis, b, m, *args, **kwargs):
    
    "Add a line with slope m and intercept b to ax"
    xval = np.linspace(df['horsepower'].min(),df['horsepower'].max(),200)
    
    axis = plt.plot(xval, b + m * xval, *args, **kwargs)
```

```python
fig, ax = plt.subplots(1,1,figsize = (7,4))

ax.set_xlabel('horsepower')
ax.set_ylabel('mpg')

ax = plt.scatter(df['horsepower'],df['mpg'], marker = 'x', color = 'blue')

lsline(ax, fit_results.params[0], fit_results.params[1], color = 'red', linewidth = 2)
```

- As the plot above suggest, there are some predictor values that can influence the fit however the highest leverage value among the predictors is way below the average leverage of all the observations. So we conclude that these points do not have much influence on the result of the fit.

#### 9.e

Lots of different couplings can be considered. For example, in the recent years the car weight and horsepower appear to be lesser and lesser. Cylinder numbers seem to affect the weight as well. Lets add all to see if the model is improved. 

```python
XMLreg_train.head()
```

```python
XMLreg_train['cylinders:weight'] = XMLreg_train['cylinders'] * XMLreg_train['weight']
```

```python
XMLreg_train['year:weight'] = XMLreg_train['year'] * XMLreg_train['weight']
```

```python
XMLreg_train['year:horsepower'] = XMLreg_train['year'] * XMLreg_train['horsepower']
```

```python
XMLreg_train.head()
```

```python
MLreg_model = sm.OLS(yMLreg_train, XMLreg_train)

MLreg_results = MLreg_model.fit()

print(MLreg_results.summary())
```

- Year-weight interaction does not appear to be statistically significant however cylinder:weight and year:horsepower interactions are statistically significant. Lets add origin:displacement, origin:horsepower, displacement:weight, displacement:horsepower AND drop accelerelation

```python
XMLreg_train['origin:horsepower'] = XMLreg_train['origin'] * XMLreg_train['horsepower']
XMLreg_train['origin:displacement'] = XMLreg_train['origin'] * XMLreg_train['displacement']
```

```python
XMLreg_train['weight:displacement'] = XMLreg_train['weight'] * XMLreg_train['displacement']
XMLreg_train['horsepower:displacement'] = XMLreg_train['horsepower'] * XMLreg_train['displacement']
```

```python
XMLreg_train.head()
```

```python
XMLreg_train = XMLreg_train.drop(['acceleration'], axis = 1)
```

```python
MLreg_model2 = sm.OLS(yMLreg_train, XMLreg_train)

MLreg_results2 = MLreg_model2.fit()

print(MLreg_results2.summary())
```

- These results indicate that the interactions that appear to be statistically significant are displacement:weight, horsepower:origin and horsepower:year.

```python
# Lets apply all the x-forms to horsepower, drop cylinders and acceleration

XMLreg_m = XMLreg

XMLreg_m['horsepowersq'] = np.power(XMLreg_m['horsepower'],2)
XMLreg_m['horsepowersqr'] = np.sqrt(XMLreg_m['horsepower'])
XMLreg_m['horsepowerlog'] = np.log(XMLreg_m['horsepower'])

XMLreg_m.drop(['acceleration','cylinders'], axis = 1, inplace = True)
```

```python
MLreg_modelm = sm.OLS(yMLreg_train, XMLreg_m_train)

MLreg_resultsm = MLreg_modelm.fit()

print(MLreg_resultsm.summary())
```
